Age Validation-Improve the program by incorporating a method called age validation() to verify that the provided age is a positive integer. If the age is deemed invalid (less than 3 or greater than 60), handle the validation accordingly.

```python
class Student:
    def __init__(self, enrollment_id, student_name, course_name, age, marks, grade=None):
        self.enrollment_id = enrollment_id
        self.student_name = student_name
        self.course_name = course_name
        self.age = age
        self.marks = marks
        self.grade = grade

    def age_validation(self):
        if not isinstance(self.age, int):
            raise ValueError("Age must be an integer")
        if self.age < 3 or self.age > 60:
            raise ValueError("Age must be between 3 and 60")
        print("Age is valid")

student1 = Student("1", "raja", "Math", 25, 85)
student1.age_validation()
```

Marks Validation - Modify the program with a method validate_and_convert_marks() to ensure that the entered marks are between 0 and 44. If the entered marks are within this range, convert them into a percentage out of 44 and return the result Otherwise, return "Invalid Marks

```python

class Student:
    def __init__(self, enrollment_id, student_name, course_name, age, marks, grade=None):
        self.enrollment_id = enrollment_id
        self.student_name = student_name
        self.course_name = course_name
        self.age = age
        self.marks = marks
        self.grade = grade

    def validate_and_convert_marks(self):
        if not isinstance(self.marks, int):
            return "Invalid Marks"
        if 0 <= self.marks <= 44:
            return (self.marks / 44) * 100  # Convert to percentage
        else:
            return "Invalid Marks"

student1 = Student("1", "raja", "Math", 25, 85)
percentage = student1.validate_and_convert_marks()
print(percentage)

```

It should then print the details of all guardians entered


c Implement a static method named 'mobile to manage mobile numbers. This method should prompt the user to amput the mobile number for each guardian. Ensure that the mobile number format consists of 7 digits and is composed antirely of numbers. Print a success message if the format is vald otherwise, print an error message

```python
class Guardian:
    guardian_count = 0

    def __init__(self):
        while True:
            try:
                self.enrollment_id = int(input("Enter enrollment ID: "))
                break
            except ValueError:
                print("Invalid input. Please enter an integer enrollment ID.")
        self.name = ""
        self.mobile_number = ""

    @classmethod
    def add_guardian(cls):
        while True:
            name = input("Enter guardian's name: ")
            cls.guardian_count += 1
            guardian = Guardian()
            guardian.name = name
            Guardian.mobile(guardian)
            print(f"Guardian {cls.guardian_count} added successfully.")

            add_more = input("Do you want to add another guardian? (yes/no): ")
            if add_more.lower() != "yes":
                break

    @staticmethod
    def mobile(guardian):
        while True:
            mobile_number = input("Enter guardian's mobile number (7 digits): ")
            if len(mobile_number) == 7 and mobile_number.isdigit():
                guardian.mobile_number = mobile_number
                print("Mobile number added successfully.")
                break
            else:
                print("Invalid mobile number format. Please enter a 7-digit number.")

# Example usage:
Guardian.add_guardian()

```
