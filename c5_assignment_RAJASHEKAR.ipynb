1. In univariate analysis, which two graphs commonly used for categorical columns. 
Explain in brief

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {"Department": ["HR", "IT", "Sales", "HR", "IT", "Sales", "HR", "Finance"]}
df = pd.DataFrame(data)
department_counts = df["Department"].value_counts()
plt.figure(figsize=(6,4))
department_counts.plot(kind="bar", color=["blue", "red", "green", "purple"])
plt.xlabel("Department")
plt.ylabel("Count")
plt.title("Employee Count by Department")
plt.show()
```

2. In univariate analysis, which two graphs commonly used for numerical columns. 
Explain in brief

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating sample numerical data (random ages)
ages = np.random.randint(18, 60, 100)

# Plotting histogram
plt.figure(figsize=(6,4))
plt.hist(ages, bins=10, color='blue', edgecolor='black')
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.title("Age Distribution of Employees")
plt.show()
```

3. Explain the difference between mean, median, and mode in the context of 
numerical columns. Provide a scenario where each measure might be more 
appropriate.

```python
import numpy as np
from scipy import stats  # Import stats for mode calculation

salaries = [40000, 45000, 50000, 55000, 60000]

# Calculating Mean, Median, and Mode
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)

mode_result = stats.mode(salaries, keepdims=True)  # Ensuring compatibility with SciPy versions
mode_salary = mode_result.mode.item()  # Convert scalar array to a single value

# Printing Results
print("Mean Salary:", mean_salary)       # Output: 50000.0
print("Median Salary:", median_salary)   # Output: 50000.0
print("Mode Salary:", mode_salary)       # Output: 40000
```

4. Explain how to interpret a box plot for a numerical column. Include discussions on 
quartiles, median, and the identification of potential outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (includes some outliers)
data = [10, 12, 14, 16, 18, 20, 30, 35, 40, 100, 120]

# Creating a box plot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot of Sample Data")
plt.xlabel("Values")
plt.show()
```

5. Compare and contrast standard deviation and variance as measures of dispersion. 
How can these measures help in understanding the variability of numerical data?

```python
import numpy as np

# Sample data
data = [10, 15, 20, 25, 30, 100]  # Includes an outlier

# Calculating variance and standard deviation
variance = np.var(data, ddof=0)  # Population variance
std_dev = np.std(data, ddof=0)  # Population standard deviation

print("Variance:", variance)  
print("Standard Deviation:", std_dev)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample numerical data (e.g., salaries in dollars)
data = [30000, 35000, 40000, 45000, 50000, 55000, 60000, 70000, 80000, 90000, 100000]

# Create a histogram with 5 bins
plt.hist(data, bins=5, edgecolor='black', color='skyblue')

# Add labels and title
plt.xlabel("Salary Range ($)")
plt.ylabel("Frequency")
plt.title("Salary Distribution Histogram")

# Show the plot
plt.show()
```
